```python
# Retirando todas as colunas que não serão importantes na análise
df.drop(["Case Number","Date","Year","Area","Location","Activity","Name","Time","Species ",'Investigator or Source','pdf','href formula','Case Number.1','Case Number.2','original order','Unnamed: 22','Unnamed: 23',"href"], inplace=True , axis=1)
```

```python
#Verificação do novo dataframe
df.head()
```

```python
# Contagem de acidentes por sexo
df['Sex '].value_counts()
```

Ainda que todos os valores inválidos para o sexo fossem considerados como femininos, 80% das vítimas dos ataques são homens. Considerando só os casos em que há documentação, 88% são homens.

```python
# Proporção dos acidentes por sexo
df['Sex '].value_counts(normalize=True)
```

```python
#Contagem dos acidentes fatais
df["Fatal (Y/N)"].value_counts()
```

```python
# Criação de um novo dataframe(pof) com somente os acidentes que são fatais
pof=df[df['Fatal (Y/N)'] == "Y"]
```

```python
#Contagem dos acidentes fatais por sexo
pof["Sex "].value_counts()
```

```python
# Acidentes fatais, por sexo, em proporção
pof["Sex "].value_counts(normalize=True)
```

No caso das vítimas fatais, a proporção masculina é ainda maior, variando entre 84% e 91%, dependendo das simplificações.
